```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
df = pd.read_csv("C:/Users/nurik/Desktop/uber.csv") #Считываем данные из csv файла в DataFrame
uber = df #создадим копию и будем работать с ней, не изменяя первоначальный набор данных
```

```python
# Проведем разведочный анализ данных EDA
#Посмотрим на первые 5 строк
print("Первые 5 строк данных:")
print(uber.head())
```

```python
#Посмотрим на статистику наших данных: число значений, среднее значение и тд. Уже видно, что есть пропуски
print("\nСтатистика данных:")
print(uber.describe())
```

```python
print("\nИнформация о данных:")
print(uber.info())
```

```python
print("\nПропущенные значения:")
print(uber.isna().sum()) #Выведем пропущенные значения
```

```python
# Удаление пропусков
uber.dropna(inplace=True)
# Построим распределение стоимости поездок
plt.figure(figsize=(10, 5))
sns.histplot(uber['fare_amount'], bins=50)
plt.title('Распределение стоимости поездок (исходные данные)')
plt.show()
```

```python
#Видим, что у некоторых пользователей отрицательная стоимость - удалим эти поездки
uber = uber[uber['fare_amount'] > 0]
#Из графика уже видно, что почти 80% клиентов попалает в небольшой интервал. Для большей точности рассчитаем перцентили:\
print("Анализ выбросов в стоимости:")
print(f"80-й перцентиль: {uber['fare_amount'].quantile(0.8):.2f}$")
print(f"99-й перцентиль: {uber['fare_amount'].quantile(0.99):.2f}$")
print(f"99.5-й перцентиль: {uber['fare_amount'].quantile(0.995):.2f}$")
print(f"99.9-й перцентиль: {uber['fare_amount'].quantile(0.999):.2f}$")
```

```python
# Удаляем выбросы в стоимости (оставляем 99% данных)
fare_upper_limit = uber['fare_amount'].quantile(0.99)
print(f"Удаляем поездки дороже ${fare_upper_limit:.2f}")
uber = uber[uber['fare_amount'] <= fare_upper_limit]
```

```python
#Теперь приведем данные к их правильному формату
uber['pickup_datetime'] = pd.to_datetime(uber['pickup_datetime'])
uber['hour'] = uber['pickup_datetime'].dt.hour
uber['day_of_week'] = uber['pickup_datetime'].dt.dayofweek
uber['month'] = uber['pickup_datetime'].dt.month
```

```python
# Анализируем среднюю стоимость поездки по часам
hourly_avg_fare = uber.groupby('hour')['fare_amount'].mean().reset_index() #cреднее по часам
plt.figure(figsize=(12, 6))
sns.lineplot(x='hour', y='fare_amount', data=hourly_avg_fare, marker='o') #задаем оси
plt.axhline(y=hourly_avg_fare['fare_amount'].mean(), color='r', linestyle='--', label='Средняя стоимость')
plt.title('Средняя стоимость поездки по часам дня') #подписываем график
plt.xlabel('Час дня')
plt.ylabel('Средняя стоимость ($)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Из графика мы видим очень интересный результат, средняя стоимость поездки в (4,5) утра существенно выше.
# Находим пиковые часы
fare_mean = hourly_avg_fare['fare_amount'].mean()
peak_hours = hourly_avg_fare[hourly_avg_fare['fare_amount'] > fare_mean * 1.2]['hour'].tolist() 
print(f"Часы с повышенной стоимостью (> +20% от среднего): {sorted(peak_hours)}")
uber['is_peak_hour'] = uber['hour'].isin(peak_hours).astype(int)
```

```python
# Создаем признак расстояния
def haversine_distance(lon1, lat1, lon2, lat2):
    R = 6371
    lon1, lat1, lon2, lat2 = map(np.radians, [lon1, lat1, lon2, lat2])
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = np.sin(dlat/2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2)**2
    c = 2 * np.arcsin(np.sqrt(a))
    return R * c
uber['distance_km'] = haversine_distance(
    uber['pickup_longitude'], uber['pickup_latitude'],
    uber['dropoff_longitude'], uber['dropoff_latitude'])
```

```python
# Анализируем выбросы в расстоянии
print("Анализ выбросов в расстоянии:")
print(f"99-й перцентиль: {uber['distance_km'].quantile(0.99):.2f} км")
print(f"99.5-й перцентиль: {uber['distance_km'].quantile(0.995):.2f} км")
```

```python
# Удаляем выбросы в расстоянии (оставляем 99% данных)
distance_upper_limit = uber['distance_km'].quantile(0.99)
print(f"Удаляем поездки длиннее {distance_upper_limit:.2f} км")
uber = uber[uber['distance_km'] <= distance_upper_limit]
```

```python
# Проверим распределение расстояний
plt.figure(figsize=(10, 5))
sns.histplot(uber['distance_km'], bins=50)
plt.title('Распределение расстояния поездок после удаления выбросов')
plt.show()
```

```python
# Проверим распределение стоимости после всех очисток
plt.figure(figsize=(10, 5))
sns.histplot(uber['fare_amount'], bins=30)
plt.title('Распределение стоимости поездок после удаления выбросов')
plt.show()
```

```python
#Проанализируем еще по нескольким данным
# Анализ по дням недели
weekday_names = ['Пн', 'Вт', 'Ср', 'Чт', 'Пт', 'Сб', 'Вс']
daily_avg_fare = uber.groupby('day_of_week')['fare_amount'].mean().reset_index()
plt.figure(figsize=(12, 6))
sns.barplot(x='day_of_week', y='fare_amount', data=daily_avg_fare)
plt.xticks(range(7), weekday_names)
plt.title('Средняя стоимость поездки по дням недели')
plt.xlabel('День недели')
plt.ylabel('Средняя стоимость ($)')
plt.show()
#Видим, что существенной разницы от дня недели нет
```

```python
# Анализ по месяцам
monthly_avg_fare = uber.groupby('month')['fare_amount'].mean().reset_index()
plt.figure(figsize=(12, 6))
sns.lineplot(x='month', y='fare_amount', data=monthly_avg_fare, marker='o')
plt.title('Средняя стоимость поездки по месяцам')
plt.xlabel('Месяц')
plt.ylabel('Средняя стоимость ($)')
plt.xticks(range(1, 13))
plt.grid(True)
plt.show()
```

```python
# Тепловая карта стоимости поездок
#Совместим зависимость от часа и дня недели 
pivot_table = uber.pivot_table(
    values='fare_amount',
    index='hour',
    columns='day_of_week',
    aggfunc='mean')
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table, cmap='YlOrRd', annot=True, fmt='.1f')
plt.title('Тепловая карта стоимости поездок (час дня × день недели)')
plt.xlabel('День недели')
plt.ylabel('Час дня')
plt.show()
```

```python
#Создадим дополнительные признаки из координат
uber['pickup_lat_round'] = round(uber['pickup_latitude'], 2)
uber['pickup_lon_round'] = round(uber['pickup_longitude'], 2)
uber['dropoff_lat_round'] = round(uber['dropoff_latitude'], 2)
uber['dropoff_lon_round'] = round(uber['dropoff_longitude'], 2)
x = uber[['hour', 'day_of_week', 'month', 'distance_km', 
          'is_peak_hour', 
          'pickup_lat_round', 'pickup_lon_round',
          'dropoff_lat_round', 'dropoff_lon_round']]
y = uber['fare_amount']
```

```python
# Подготовка данных для машинного обучения
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics

# Подготовка признаков и целевой переменной
x = uber[['hour', 'day_of_week', 'month', 'distance_km', 'is_peak_hour']]
y = uber['fare_amount']

# Разделяем данные на обучающую и тестовую выборки
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)

# Линейная регрессия
lrmodel = LinearRegression()
lrmodel.fit(x_train, y_train)
lr_predict = lrmodel.predict(x_test)

# Оценка модели линейной регрессии
print("Линейная регрессия:")
print(f"Средняя абсолютная ошибка: {metrics.mean_absolute_error(y_test, lr_predict):.2f}")
print(f"Среднеквадратичная ошибка: {metrics.mean_squared_error(y_test, lr_predict):.2f}")
print(f"Корень из среднеквадратичной ошибки: {np.sqrt(metrics.mean_squared_error(y_test, lr_predict)):.2f}")
print(f"R² score: {metrics.r2_score(y_test, lr_predict):.4f}")
```

```python
# Случайный лес
rfmodel = RandomForestRegressor(n_estimators=100, random_state=101)
rfmodel.fit(x_train, y_train)
rf_predict = rfmodel.predict(x_test)

# Оценка модели случайного леса
print("\nСлучайный лес:")
print(f"Средняя абсолютная ошибка: {metrics.mean_absolute_error(y_test, rf_predict):.2f}")
print(f"Среднеквадратичная ошибка: {metrics.mean_squared_error(y_test, rf_predict):.2f}")
print(f"Корень из среднеквадратичной ошибки: {np.sqrt(metrics.mean_squared_error(y_test, rf_predict)):.2f}")
print(f"R² score: {metrics.r2_score(y_test, rf_predict):.4f}")
```

```python
# Сравнение моделей
print("\nСравнение моделей:")
print(f"Линейная регрессия R²: {metrics.r2_score(y_test, lr_predict):.4f}")
print(f"Случайный лес R²: {metrics.r2_score(y_test, rf_predict):.4f}")
```

```python
# Анализ важности признаков для случайного леса
feature_importance = pd.DataFrame({
    'feature': x.columns,
    'importance': rfmodel.feature_importances_
}).sort_values('importance', ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x='importance', y='feature', data=feature_importance)
plt.title('Важность признаков в модели случайного леса')
plt.show()

# Визуализация предсказаний vs реальных значений
plt.figure(figsize=(12, 5))


# Для линейной регрессии
plt.subplot(1, 2, 1)
plt.scatter(y_test, lr_predict, alpha=0.3)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
plt.xlabel('Реальная стоимость')
plt.ylabel('Предсказанная стоимость')
plt.title('Линейная регрессия')

# Для случайного леса
plt.subplot(1, 2, 2)
plt.scatter(y_test, rf_predict, alpha=0.3)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
plt.xlabel('Реальная стоимость')
plt.ylabel('Предсказанная стоимость')
plt.title('Случайный лес')

plt.tight_layout()
plt.show()
```
